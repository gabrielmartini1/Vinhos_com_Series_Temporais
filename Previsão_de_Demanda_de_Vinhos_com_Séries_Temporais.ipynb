##**1. Introdução**

<p align="justify">O vinho é uma bebida alcoólica produzida a partir da fermentação do suco de uvas. Cada país tem suas próprias variedades de uvas, técnicas de produção e tradições culturais associadas ao vinho.</p>

<p align="justify">Os vinhos podem ser classificados em várias categorias, dependendo de diferentes fatores, como o tipo de uva usada, o método de produção, o local de origem e o tempo de envelhecimento. Alguns exemplos de categorias de vinho incluem vinho branco, vinho tinto, vinho rosé, vinho espumante e vinho fortificado.</p>

<p align="justify">O vinho é apreciado por muitas pessoas em todo o mundo, tanto por seu sabor único como por suas propriedades benéficas para a saúde, quando consumido com moderação. O consumo moderado de vinho tem sido associado a uma série de benefícios para a saúde, incluindo a redução do risco de doenças cardiovasculares, colesterol e a melhoria da função cognitiva.</p>

<p align="justify">Além de seu valor como uma bebida apreciada, o vinho também é um produto econômico importante em muitos países. A indústria do vinho emprega milhões de pessoas em todo o mundo e gera bilhões de dólares em receita anualmente.</p>

<p align="justify">O mercado mundial de vinhos tem enfrentado mudanças significativas nas últimas décadas. Estas mudanças seriam fruto do aumento de competitividade dos países exportadores do hemisfério sul, principalmente do Chile, a partir dos anos 1990, e da Argentina, na última década.</p>

<p align="justify">Além de Chile e Argentina, outros países se tornaram importantes no mercado internacional do produto, como EUA, Austrália e África do Sul, experimentando aumento da sua representatividade nas exportações mundiais. Como resultado, houve queda no preço médio dos vinhos comercializados no mercado internacional, além de maior variedade de produtos disponíveis ao consumidor.</p>

<p align="justify">Segundo informações da International Organization of Vine and Wine (OIV), o consumo per capita de vinho no Brasil foi de 1,84 litro em 2011. Este nível de consumo pode ser considerado baixo quando comparado com países europeus, como França (46 litros), Portugal (44 litros) e Itália (38 litros). Mesmo quando comparado ao consumo de alguns países sul-americanos, o consumo per capita de vinho no Brasil também pode ser considerado baixo. O consumo de vinho da Argentina, do Uruguai, Chile e Paraguai foi 13, 11, 10 e 3 vezes, respectivamente, maior que o brasileiro em 2011.</p>

<p align="justify">Atualmente, o consumo de vinho nos países europeus ainda é relativamente alto, embora tenha havido algumas mudanças nos últimos anos. De acordo com dados de 2019 do Observatório Mundial de Vinho, o consumo per capita de vinho na França foi de cerca de 44 litros, o consumo em Portugal foi de cerca de 51 litros e o consumo na Itália foi de cerca de 29 litros.</p>

<p align="justify">No entanto, é importante notar que esses números podem variar de ano para ano e também dependem de fatores como a situação econômica, as tendências de consumo e as políticas de saúde pública em relação ao consumo de álcool. Além disso, muitos países estão experimentando um aumento no consumo de vinhos de países como Chile, Argentina, Austrália e Estados Unidos.</p>

<p align="justify">Em um mercado competitivo as empresas devem procurar a satisfação do cliente fornecendo o produto no momento solicitado. Para isso deve-se ter a habilidade de realizar tomadas de decisões rápidas e precisas no seu planejamento, com isso a companhia deve ter uma habilidade especial em extrair e analisar certos dados relevantes que possam dar direcionamento às tomadas de decisão.</p>

<p align="justify">A previsão de demanda é uma ferramenta que pode auxiliar a essas empresas que buscam adotar um planejamento estratégico e compreender o comportamento dos consumidores para aumentar, crescer ou explorar mercados novos, pois este permite que as organizações conheçam o que o seu público deseja, o momento e a quantidade requerida. As técnicas de previsão de demanda possibilitam ao interessado extrair previsões futuras que são relevantes para as tomadas de decisões utilizando dados da demanda anterior.</p>

<p align="justify">Para uma empresa é importante ter conhecimento sobre a demanda do seu produto no mercado, com ela é possível determinar o comportamento do público ante ao consumo de bens e serviços fornecidos. Com a previsão de demanda é possível traçar rumos em relação às atividades que serão realizadas com as informações oferecidas. A previsão da demanda é essencial para as empresas principalmente na área do planejamento estratégico da produção, pois com o auxílio desta ferramenta é possível realizar planos de capacidade, fluxo de caixa, de vendas, de produção e de estoque.</p>

<p align="justify">Cumprindo igualmente uma função importante nos processos de planejamento dos sistemas de produção porque através dos dados fornecidos é possível que os encarregados dos setores prevejam o futuro e planejem corretamente suas ações. A incerteza da demanda do produto/serviço no mercado causa uma instabilidade no setor do supply chain para a realização das compras com o fornecedor, com isto pode se observar que a previsão da demanda impacta positivamente no setor.</p>

<p align="justify">A previsão de demanda é comumente utilizada no comercio como uma tarefa estatística, pois pode auxiliar na tomada de decisões em relação à programação da produção, a logística, a quantidade de pessoal envolvido e com isso podendo guiar o planejamento estratégico para prazos longos. Deve ser considerando como incondicional para as atividades que requerem a tomada de decisões dentro da gestão, podendo impactar distintas áreas das empresas. As previsões de demanda podem ser classificadas de curto, médio e longo prazo dependendo das necessidades da empresa.</p>

## **4. Pré-processamento de Dados**

<p align="justify">O pré-processamento de dados é uma etapa fundamental em todos os projetos de ciências de dados e engenharia de machine learning, onde os dados brutos são transformados em um formato que possa ser facilmente analisado e modelado por algorítmos de aprendizado de máquina.</p>

<p align="justify">Esta etapa de pré-processamento de dados envolve a execução de uma série de tarefas, tais como: a limpeza dos dados, redução, integração de dados, criação ou transformação de novas variáveis, seleção de dados, balanceamento de classes, divisão do conjunto de dados, padronização e normalização dos dados.</p>

<p align="justify">A limpeza envolve a identificação e remoção de dados duplicados, inconsistências, erros e outliers, enquanto a transformação pode incluir normalização, discretização e codificação de variáveis categóricas.</p>

<p align="justify">Redução de dimensionalidade é uma técnica para reduzir a complexidade do modelo, como por exemplo, através da seleção de características relevantes e descarte de variáveis redundantes para melhorar a performance do modelo.</p>

<p align="justify">A integração envolve a combinação de dados de múltiplas fontes em um único conjunto de dados. Já a seleção de dados é a escolha das variáveis mais relevantes para o modelo preditivo.</p>

<p align="justify">A criação ou a transformação de novas variáveis (features engineering), ocorre a partir das variáveis existentes, para que possam ter maior relevância para o modelo.</p>

<p align="justify">O balanceamento ocorre quando a classe alvo do modelo está desbalanceada. É necessário aplicar técnicas para balanceá-la, a fim de evitar um modelo enviesado.</p>

<p align="justify">A divisão do conjunto de dados, refere-se a separação dos dados em treino, validação e teste, a fim de avaliar o modelo de forma adequada.</p>

<p align="justify">A padronização e a normalização são técnicas de "feature scaling". Essas técnicas são usadas para colocar as variáveis (ou features) em uma mesma escala numérica. O objetivo da "feature scaling" é garantir que todas as variáveis tenham um peso igual no modelo de machine learning.</p>

<p align="justify">O objetivo do pré-processamento de dados é melhorar a qualidade dos dados, prepará-los para serem usados em modelos de aprendizado de máquina e garantir que os resultados da análise sejam precisos e confiáveis. A qualidade dos dados tem um impacto direto na eficácia dos modelos e na tomada de decisões baseadas nos resultados da análise de dados.</p>

###**4.1.3 Valores inconsistentes**

<p align="justify">Analisando o dataset "products_csv", observa-se que a instância 'vintage' apresenta alguns valores classificados como NV (Non-Vintage), os quais serão substituídos por zero.</p>

<p align="justify">A instância 'price_brl' apresenta alguns valores com vírgulas como separador de milhares. Vamos remover as vírgulas dos valores do Real para que eles fiquem no mesmo formato que os valroes em Dólar.</p>

<p align="justify">Para realizarmos essa alteração, vamos usar o método "str.replace( )" do pandas para remover as vírgulas da coluna 'price_brl', e em seguida converteremos os valores para ponto flutuante usando o método "astype( )".
```python
# Remover NV da instância 'vintage'
df_wines.vintage = df_wines.vintage.replace('NV', '0')
df_wines.vintage = pd.to_numeric(df_wines.vintage)

# Excluir as vírgulas de 'price_brl'
df_wines.replace(',','', regex=True, inplace=True)
```

### **5.5.1 Calculando o total em vendas por item**

<p align="justify">Para calcularmos o total em vendas por item, será necessário gerarmos 2 novas features no DataFrame (df_income), 'total_brl' e 'total_usd'. Para a feature 'total_brl', vamos multiplicar 'price_brl' por 'sales' e para a feature 'total_usd', vamos multiplicar as instâncias 'price_usd' por 'sales', conforme abaixo.</p>

### **6.2 Testes Estatísticos e ARIMA**

<p align="justify">Os testes estatísticos e ARIMA são 2 ferramentas importantes para a análise de séries temporais. Os testes estatísticos são utilizados para verificar a estacionariedade da série temporal.</p>

<p align="justify">Uma série temporal é considerada estacionária se suas propriedades estatísticas, como média e variância, são constantes ao longo do tempo. A estacionariedade é importante para a aplicação de modelos de previsão, como o ARIMA. Alguns testes estatísticos comuns incluem o teste de Dickey-Fuller aumentado (ADF) e o teste de Kwiatkowski-Phillips-Schmidt-Shin (KPSS).</p>

<p align="justify">Já o modelo ARIMA (AutoRegressive Integrated Moving Average) é um modelo matemático que combina componentes autoregressivos (AR) e médias móveis (MA) para modelar e prever séries temporais. O modelo ARIMA assume que a série temporal é estacionária e que a relação entre as observações em diferentes tempos é linear. Para ajustar um modelo ARIMA, é necessário identificar os parâmetros p, d e q, que representam o número de termos autoregressivos, o número de diferenças e o número de termos de média móvel, respectivamente.</p>

<p align="justify">Os testes estatísticos são importantes para garantir a adequação do modelo ARIMA para a série temporal em análise. Se a série não for estacionária, é necessário aplicar técnicas de diferenciação para torná-la estacionária antes de ajustar o modelo ARIMA. Além disso, a escolha dos parâmetros p, d e q pode ser realizada por meio de testes estatísticos como o teste de Ljung-Box, que avalia se os resíduos do modelo são aleatórios e independentes, indicando uma boa escolha dos parâmetros.</p>

<p align="justify">Para verificarmos se uma série temporal é estacionária, é possível realizarmos um conjunto de testes estatísticos, tais como:</p>

* <p align="justify">Análise visual: Observar o gráfico da série temporal e verificar se há presença de tendência e/ou sazonalidade. Caso haja, a série não é estacionária.</p>

* <p align="justify">Estatísticas de resumo: Calcular as estatísticas de resumo da série temporal, como média e variância, em janelas de tempo distintas. Caso essas estatísticas variem significativamente entre as janelas de tempo, a série não é estacionária.</p>

* <p align="justify">Teste de Dickey-Fuller Aumentado (ADF): É um teste estatístico que tem como hipótese nula a presença de raiz unitária na série temporal. Caso a hipótese nula seja rejeitada, a série é considerada estacionária.</p>

* <p align="justify">Teste de Kwiatkowski-Phillips-Schmidt-Shin (KPSS): É um teste estatístico que tem como hipótese nula a estacionariedade da série temporal. Caso a hipótese nula seja rejeitada, a série não é estacionária.</p>

### **6.2.1 Teste ADF**

<p align="justify">O ADF é um dos testes estatísticos mais usados para TS. Suas hipóteses nula e alternativa são:</p>

* <p align="justify">A Hipótese Nula ( 𝐻0 ) do teste é que a TS não é estacionária. Ou seja, possui algum tipo de dependência em relação ao tempo.</p>

* <p align="justify">A Hipótese Alternativa ( 𝐻1 ) rejeita a hipótese nula, ou seja, que a TS é estacionária.</p>

### **6.2.2 Preparando o Teste ADF**

<p align="justify">Para utilizarmos o Prophet, será necessário padronizarmos as features 'date' e 'sales' do DataFrame "df_income". A coluna 'date' deverá ser renomeada para 'ds' e a coluna 'sales' será renomeada para 'y', conforme mostra o código abaixo:</p>

```python
# Fazendo uma cópia do DataFrame "df_income"
df_ts = df_income[['date', 'sales']].copy()

# Agrupar os dados por data e calcular a soma das vendas diárias
df_ts = df_income.groupby('date', as_index=False)['sales'].sum()

# Renomear as colunas para o padrão de séries temporais
df_ts = df_ts.rename(columns={'date': 'ds', 'sales': 'y'})

# Visualizar as primeiras linhas do DataFrame resultante
df_ts.head()
```

### **6.2.2 Aplicando o Teste ADF**

<p align="justify">Utilizando o pacote statsmodels e a função adfuller, vamos aplicar o teste de Augmented Dickey Fuller (ADF) na série temporal representada pels valores X. Lembrando que o Teste ADF é utilizado para verificar se uma série temporal é estacionária ou não. A hipótese nula do teste é que a série temporal possui uma raiz unitária, indicando que não é estacionária. O valor de teste (Statistical Test) é comparado com os valores críticos (Critical Values) para determinar se a série é estacionária. A seguir, vamos apresentar o código e o output do Teste ADF realizado:</p>

```python
# importanto o pacote
from statsmodels.tsa.stattools import adfuller

# extraindo os valores
X = df_ts.y

# aplicando o ADF e imprimindo os resultados.
result = adfuller(X)

print("--"*14)
print("Augmented Dickey–Fuller".center(28).upper())
print("--"*14)
print('Estatística do teste: {:.4f}'.format(result[0]))
print('Valor(p): {:.10f}'.format(result[1]))
print('Valores Críticos:')
for key, value in result[4].items():
	print('\t{}: {:.4f}'.format(key, value))
```

<p align="justify">Observa-se que o valor de teste é -2.3601 e o Valor (p) é 0.1533303754. Como o Valor (p) é maior do que o nível de significância comumente usado de 0.05, não há evidência suficiente para rejeitarmos a hipótese nula de que a série é do tipo não estacionária. Portanto, a série temporal não é considerada estacionária com base nesse teste. Os valores críticos são fornecidos para diferentes níveis de significância: 1%, 5% e 10%.</p>

<p align="justify">Sabendo que a estacionariedade de uma série temporal pode ter implicações significativas na análise e modelagem dos dados, e considerando nosso objetivo de realizar previsões precisas, é necessário realizar uma transformação na série para torná-la estacionária antes de prosseguirmos.</p>

### **6.2.3 Transformando a Série em Estacionária**

<p align="justify">A transformação de uma série temporal não estacionária em uma série estacionária pode ser feita de diversas maneiras, conforme apresentado abaixo:</p>

* <p align="justify">Diferenciação: Uma das maneiras mais simples é aplicar a diferenciação na série temporal. Isso envolve subtrair o valor atual do valor anterior em cada observação. A diferenciação de primeira ordem é comumente utilizada para tornar a série estacionária.</p>

* <p align="justify">Transformação logarítmica: Se a série apresenta um comportamento exponencial ou tendência não linear, uma transformação logarítmica pode ajudar a linearizar os dados e torná-los estacionários.</p>

* <p align="justify">Remoção de tendência: Se a série apresentar uma tendência linear, podemos removê-la, ajustando uma linha de regressão aos dados e subtraindo-a da série original.</p>

* <p align="justify">Detrending: Outra abordagem é utilizar métodos estatísticos para remover a tendência dos dados, como a decomposição aditiva ou multiplicativa, que separam os componentes de tendência, sazonalidade e aleatoriedade.</p>

* <p align="justify">Remoção de sazonalidade: Se a série temporal exibir padrões sazonais, é possível remover essa sazonalidade aplicando a diferenciação sazonal, ou seja, subtraindo os valores de um mesmo período em ciclos diferentes.</p>

### **6.2.3.1 Técnica de Detrending**

<p align="justify">Para este projeto, vamos optar inicilamente pela técnica de Detrending, que é o processo de remoção da tendência presente nos dados, visando isolar as flutuações não sistemáticas ou as variações residuais.

<p align="justify">Se a série detrending apresentar um padrão semelhante em termos de tamanho e forma da série original, isso indica que a tendência linear foi efetivamente removida dos dados. A remoção da tendência linear significa que as flutuações em torno da tendência foram eliminadas, resultando em uma série que agora reflete as variações residuais. A seguir, vamos calcular a tendência linear e compará-la no gráfico com a Série Original e a Série Detrending.</p>

```python
# Calculando a tendência linear
time = np.arange(len(df_ts))
trend = np.polyfit(time, df_ts['y'], deg=1)
trend_line = np.polyval(trend, time)

# Obtendo a série detrended
detrended_series = df_ts['y'] - trend_line

# Plotando os resultados
plt.figure(figsize=(10, 6))
plt.plot(df_ts['ds'], df_ts['y'], label='Série Original')
plt.plot(df_ts['ds'], trend_line, label='Tendência Linear')
plt.plot(df_ts['ds'], detrended_series, label='Série Detrending')
plt.legend()
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Detrending')
plt.show()
```

<p align="justify">Pela análise gráfica podemos inferir, que a série detrending apresentou um padrão semelhante em termos de tamanho e forma da série original, indicando que a tendência linear foi efetivamente removida dos dados.</p>

<p align="justify">Essa observação sugere que a série original não possui uma tendência linear significativa e as flutuações nos dados estão mais relacionadas a outros fatores, como padrões sazonais, tendências não lineares ou ruído aleatório. No entanto, é importante lembrarmos que a análise da série detrending deve ser complementada por outras técnicas e métodos estatísticos para uma análise mais aprofundada. É possível que existam outros padrões ou fatores subjacentes na série original que não foram capturados apenas pela remoção da tendência linear.</p>

### **6.2.3.2 Técnica da Diferenciação Sazonal**

<p align="justify">Para explorarmos a presença de padrões sazonais na série, podemos realizar análises adicionais, incluindo a aplicação da diferenciação sazonal. Essa técnica envolve a subtração dos valores da série em intervalos correspondentes a diferentes períodos sazonais. Através desse processo, busca-se eliminar os efeitos da sazonalidade, contribuindo para a obtenção de uma série estacionária.</p>

```python
# Copiando o DataFrame original
df_ds = df_ts.copy()

# Definindo a frequência da sazonalidade (por exemplo, sazonalidade diária = 1)
frequencia_sazonalidade = 1

# Realizando a diferenciação sazonal
df_ds['diferenciacao_sazonal'] = df_ds['y'] - df_ds['y'].shift(frequencia_sazonalidade)

# Removendo os valores NaN resultantes da diferenciação
df_ds = df_ds.dropna()

# Exibindo a série diferenciada sazonalmente
print(df_ds)
```

<p align="justify">Após aplicarmos a técnica de Diferenciação Sazonal no DataFrame "df_ts", o output gerado apresenta uma nova coluna chamada "diferenciacao_sazonal" contendo os resultados da técnica aplicada. Cada valor nessa coluna representa a diferença entre o valor da série original ('y') e o valor com a frequência sazonal anterior.</p>

<p align="justify">Por exemplo, se observarmos a segunda linha, o valor de "diferenciacao_sazonal" é -1082.0, o que significa que houve uma diminuição de 1082 unidades em relação ao valor anterior da série, ou seja, esses valores representam as flutuações ou variações residuais, após remover os efeitos da sazonalidade.

### **6.2.3.3 Técnica da Transformação Logarítmica**

###**6.5 Métricas de Desempenho**

<p align="justify">Ao avaliarmos a qualidade das previsões, é importante utilizarmos as métricas de desempenho, que nos ajudam a entender o quão bem nossas previsões estão se aproximando dos valores reais. Duas métricas comumente utilizadas são o Mean Absolute Percentage Error (MAPE) e o Mean Absolute Error (MAE).</p>

<p align="justify">O MAPE é uma métrica que nos indica a porcentagem média em que nossas previsões diferem dos valores reais. Ele é calculado pela média dos valores absolutos das diferenças entre as previsões e os valores reais, dividido pelo valor real e multiplicado por 100. O MAPE é útil para avaliar a precisão das previsões em termos percentuais, permitindo que tenhamos uma noção clara da magnitude dos erros em relação aos valores reais.</p>

<p align="justify">O MAE, por sua vez, é uma métrica que nos indica o valor absoluto médio do erro na previsão em relação à série real. Ele é calculado pela média dos valores absolutos das diferenças entre as previsões e os valores reais. O MAE fornece uma medida direta do erro médio das previsões em unidades da série original, permitindo uma compreensão mais intuitiva do desempenho do modelo.</p>

<p align="justify">Essas métricas são importantes para avaliarmos o desempenho do modelo de previsão e identificarmos possíveis áreas de melhoria. Podemos comparar os valores do MAPE e MAE ao longo do tempo ou em diferentes modelos para avaliar qual abordagem oferece as previsões mais precisas.</p>

```python
from sklearn.metrics import mean_absolute_error

test_md = df_log[-prediction_size:]
y_pred = forecast['yhat'][-prediction_size:].values
y_true = test_md['y'].values

def mean_absolute_percentage_error(y_true, y_pred):
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

mae = mean_absolute_error(y_true, y_pred)
mape = mean_absolute_percentage_error(y_true, y_pred)

print("MAE:  ", mae)
print("MAPE:  ", mape)
```

<p align="justify">Com base nos resultados obtidos das métricas de desempenho, concluímos que o MAE (Mean Absolute Error) encontrado foi de 0.0215, indicando que, em média, as previsões têm uma diferença absoluta de 0.0215 em relação aos valores reais. Já o MAPE (Mean Absolute Percentage Error) foi de 0.2013, o que indica que, em média, as previsões têm uma diferença de aproximadamente 0.2013% em relação aos valores reais.</p>

<p align="justify">No geral, um MAE e MAPE baixos indicam um bom desempenho do modelo, pois significa que as previsões estão próximas dos valores reais. Entretanto, ao fazermos a transformação para a série não estacionária, os dados foram colocados em uma outra escala (log).
